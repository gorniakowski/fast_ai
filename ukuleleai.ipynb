Zabieramy się za obróbkę danych.

```
np.random.seed(666) #tymczasowo robimy liczby nielosowe, żeby model pracował na tych samych grupach danych
data = ImageDataBunch.from_folder('data/', valid_pct=0.2, ds_tfms=get_transforms(), size=224, num_workers=4).normalize(imagenet_stats)
```
